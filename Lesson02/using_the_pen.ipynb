# The `pen` Class 

The `pen` class is inspired by the [Logo programming language](https://en.wikipedia.org/wiki/Logo_(programming_language). It lets you draw lines and shapes by moving a pen that looks like a turtle around a canvas. This notebook will show you all of the things you can do with the pen. Before the pen can used you have to run the code in the cell below:

```python
from Lab.drawing import pen
```

You only have to run the `import` cell once unless you restart the python kernel. 

## Drawing Lines 

The `draw` function of the pen draws a line. After you're done drawing the `show` function shows your drawing. The cell below draws a line.

```python
pen.draw(20)
pen.show()
```

You can also draw in reverse.

```python
pen.draw(50) # Go forward 50 steps.
pen.draw(-20) # Go back 20 steps.
pen.show()
```

## Turns

The `turn` function turns the turtle. The number provided to the `turn` function is the angle to turn in degrees. A positive angle is a left turn and a negative angle is a right turn. 

```python
pen.draw(30)
pen.turn(45) # Turn left
pen.draw(30)
pen.turn(-45) # Turn right
pen.draw(30)
pen.show()
```

## Picking Up and Putting Down the Pen 

You can control the drawing by picking up the pen. When the pen is up moves don't draw a line. When you want to draw lines again you can put the pen back down. The `up` and `down` functions control the pen.

```python
pen.draw(30)
pen.up()      # Pick up the pen
pen.draw(20)  # There will be no line from this draw.
pen.down()    # Put down the pen
pen.draw(30)
pen.show()
```

## The Color and Width of the Pen 

The You can control the color and the thickness of the lines that the pend draws using the `color` and `width` functions. 

```python
pen.width(30)      # Draw a 30px line.  
pen.color('red')   # Draw red
pen.draw(90)
pen.turn(120)
pen.color('green') # Draw green
pen.draw(90)
pen.turn(120)
pen.color('blue')  # Draw blue
pen.draw(90)
pen.turn(120)
pen.show()
```

The pen can use any colors that are valid in HTML. You can see HTML color codes on [this page](https://htmlcolorcodes.com/).

```python
pen.width(30)         # Draw a 30px line.  
pen.color('fuchsia')  # Draw a fuchsia 
pen.draw(90)
pen.turn(120)
pen.color('#ff1010')  # Use #RRGGBB notation. 
pen.draw(90)
pen.turn(120)
pen.color('#00ff00')  
pen.draw(90)
pen.turn(120)
pen.show()
```

## Moving the Pen to Absolute Coordinates 

The `goto` function moves the pen to a particular location. If the pen is down it will draw a line on the path to that location. If the pen is up no line will be drawn. 

```python
# The pen starts at (0, 0) facing up.

pen.draw(50)     # The pen is now at (0, 50)
pen.goto(20, 0)
pen.draw(50)     # The pen is now at (20, 50)
pen.goto(40, 0)
pen.draw(50)     # The pen is now at (40, 50)
pen.show()
```

## Controlling the Pen's Heading 

You can turn the pen to a particular direction in degrees using the `heading` funciton. 

```python
# The pen starts at (0,0), heading 90

pen.heading(0)    # Face right 
pen.draw(100)
pen.goto(0,0)
pen.heading(90)   # Face up
pen.draw(100)
pen.goto(0,0)
pen.heading(180)  # Face left
pen.draw(100)
pen.goto(0,0)
pen.heading(270)  # Face down
pen.draw(100)
pen.show()
```

## Setting a Background Image 

You can set a custom background image for the arena. The image is automatically scaled to the size of the arena. 

```python
pen.background('maze.png')
pen.show()
```

## Controlling the Shape of the Arena 

The `show` function takes arguments that control how the final plot is rendered. The default arena is square with `x = [-100, 100]` and `y = [-100, 100]`. You can control the shape of the arena by passing the `arena` parameter to `show`. The arena parameter takes a tuple in the form `(x_minimum, x_maximum, y_minimum, y_maximum)`. For example, to confine the arena to just the positive x and y coordinate space:

```python
pen.draw(20)
pen.turn(-120)
pen.draw(20)
pen.turn(-120)
pen.draw(20)
pen.show(arena=(0,30,0,30))
```

## Controlling the Final Image

The `show` function takes arguments that control how the final plot is rendered. The `size` parameter controls the size of the plot. The size parameter takes a tuple that is in the form of `(x_inches, y_inches)`. Here's how you can make a small plot.

```python
for _ in range(4):
    pen.draw(50)
    pen.turn(90)

pen.show(size=(2,2)) # Plots a 2" by 2" plot.
```

```python
for _ in range(3):
    pen.turn(120)
    for _ in range(5):
        pen.draw(70)
        pen.turn(144)

pen.show(turtle=False) # No turtle in the output
```
