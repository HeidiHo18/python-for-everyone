```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def dostar(angle):
    for _ in range(5):
        pen.draw(100)
        pen.turn(angle)
    pen.show(turtle=False)

interact(dostar, angle=widgets.IntSlider(min=0, max=360, step=1, value=144))

#dostar(288)
```
