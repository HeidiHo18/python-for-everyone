## A Flask Program 

The Lesson03 directory has a program written in [Flask](http://flask.pocoo.org/). Flask lets you write simple web applications. The flask program has two parts:

| File | Purpose | 
| --- | --- | 
| `Lesson03/web_calc/web_calc.py`| The Python program that uses Flask to generate web pages. | 
| `Lesson03/web_calc/templates/index.html` | A template file for generating HTML using the [Jinja2](http://jinja.pocoo.org/docs/2.10/) engine. | 

## Running a Flask Program 

Try out your Flask program by starting it on the command line: 

```bash
$ cd Lesson03/web_calc
$ python3.6 web_calc.py 
```

If you started the program correctly you should see output similar to this: 

```
 * Serving Flask app "web_calc" (lazy loading)
 * Environment: production
   WARNING: Do not use the development server in a production environment.
   Use a production WSGI server instead.
 * Debug mode: on
 * Running on http://0.0.0.0:8080/ (Press CTRL+C to quit)
 * Restarting with stat
 * Debugger is active!
 * Debugger PIN: 331-385-481
```

The program doesn't return to the prompt because it's waiting to serve web requests. You can access the program by pressing the "Preview" link near the "Run" link in Cloud9. The "Preview" button opens a tab in Cloud9 IDE that works like a web browser. If you would like to use your desktop browser copy the URL from the preview into your browser.

## Viewing Your Work 

When you save `web_calc.py` the Flask server reloads your program. If you have a syntax error Flask will quit and you won't be able to reload the page. Check the command line to ensure flask is running and restart it if necessary.
